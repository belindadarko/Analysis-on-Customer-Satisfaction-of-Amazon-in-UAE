```python
# Define ' ' as a null value in data set
# Import a new dataset to check null values
amazon_dataset_for_null = pd.read_excel("./dataset/amazon_dataset.xlsx",
                                        sheet_name = "original_data",
                                        header = 1,
                                        names = headers,
                                        na_values=' ')

# Checking data set with null values for basic information
print(f"""
Check basic information
{'*' * 100 }""")

print(amazon_dataset_for_null.info())


# Checking for missing values
print(f"""
Check for missing values for any variables
{'*' * 100 }

All Missing Value Counts:
------------------------------

{amazon_dataset_for_null.isnull().sum(axis = 0)}

""")
```

```python
# Creating a new column for the variable Average amount spent per visit last 3 months
amazon_dataset_for_null["y"] = amazon_dataset_for_null.loc[ : , "Average amount spent per visit last 3 months"]

# Subset the original dataset for y variable for Average amount spent per visit last 3 months > 543 AED
amazon_dataset_high = amazon_dataset_for_null.loc[ : , ["Monthly personal income"]] \
                      [amazon_dataset_for_null.loc[: , "Y >1000 AED"] == 1]

# Subset the original dataset for y variable for Average amount spent per visit last 3 months < 543 AED
amazon_dataset_low = amazon_dataset_for_null.loc[ : , ["Monthly personal income"]] \
                    [amazon_dataset_for_null.loc[ : , "Y >1000 AED"] == 0]

# Create a list that have mode, and make sure the mode is a string
# Going to replace missing values with mode

mode_high = amazon_dataset_high[["Monthly personal income"]].mode()
mode_low = amazon_dataset_low[["Monthly personal income"]].mode()

# Mode for Average amount spent per visit last 3 months > 1000 AED
print(f"""
Mode for Average amount spent per visit last 3 months > 1000 AED 
{'*' * 100 }""")

print(mode_high)

# Mode for Average amount spent per visit last 3 months < 1000 AED
print(f"""
Mode for Average amount spent per visit last 3 months < 1000 AED 
{'*' * 100 }""")

print(mode_low)

# Create a loop to input the different mode for Monthly personal income column
for element in amazon_dataset_for_null["Y >1000 AED"]:
    if element == 1:
        amazon_dataset_for_null["Monthly personal income"].fillna(mode_high.iloc[0,0], inplace=True)
    else:
        amazon_dataset_for_null["Monthly personal income"].fillna(mode_low.iloc[0,0], inplace=True)

print(amazon_dataset_for_null)
```

```python
# Creating bins for the rating variables
bins_ratings =  ["Product quality", "Service quality", "Price and quality relation", \
           "e-Commerce satisfaction rating", \
           "Likelihood of buying from e-Commerce company rating", \
           "Likelihood of recommending e-Commerce to friends rating", \
        "Variety of products offered rating", \
           "Variety of products meeting your needs rating", \
           "Navigate webpage easily rating", \
           "Easy to find products in e-Commerce app/website rating", \
           "Availability of products rating", \
           "Promotions attractiveness on app/website rating", \
           "Quality of information of product rating on app/website", \
           "Easy to compare products in e-Commerce app/website rating", \
           "Easy to manage shopping cart rating", "Experience checking-out rating", \
           "Security of app/website rating", "Information about delivery rating", \
           "Variety of delivery options satisfaction rating", \
           "Tracking package/order rating", "Time to deliver package satisfaction rating", \
           "Quality of products received rating", "Availability of customer service rating", \
           "Return and exchange process satisfaction rating"]

    
bins_sizes_ratings = [1, 4, 7, 10]
bins_ratings_labels = ["Low rating", "Middle rating", "High rating"]

for col in bins_ratings:
    amazon_dataset_for_null[f"Bins_{col}"] = pd.cut(amazon_dataset_for_null[col], \
                                                        bins = bins_sizes_ratings,
                                                    labels = bins_ratings_labels)
    
    
# Creating bins for Age
bins_age = [18, 25, 35, 45, 55, 66]

bins_age_labels = [0, 1, 2, 3, 4]

amazon_dataset_for_null["Age"] = pd.cut(amazon_dataset_for_null["Age"],
                                       bins = bins_age,
                                       labels = bins_age_labels)


# Creating bins for Number of times purchased in the last 6 months
bins_purchases = [1, 3, 6, 9]

bins_purchases_labels = [0, 1, 2]

amazon_dataset_for_null["Number of times purchased in the last 6 months"] = pd.cut(amazon_dataset_for_null["Number of times purchased in the last 6 months"],
                                       bins = bins_purchases,
                                       labels = bins_purchases_labels)



amazon_dataset_for_null.head(n=5)

    
```
